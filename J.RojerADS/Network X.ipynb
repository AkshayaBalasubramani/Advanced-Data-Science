### Individual node info

```python
#degree centality-greater the no of nodes mod=re is the degree centrality
#closeness centrality-avg distance based to get to all other nodes(connection of imp nodes)
#eigen vector centrality-the quality of the connectivity
#betweeness centrality-how much percent of the shoretest paths on graph run through u
```

```python
print(f'degree centrality={nx.degree_centrality(G9)}')
```

```python
print(f'eigen vector centrality={nx.eigenvector_centrality(G9)}')
```

```python
print(f'betweenness centrality={nx.betweenness_centrality(G9)}')
```

```python
K1=nx.complete_graph(5)
K2=nx.complete_graph(5)
K2=nx.relabel_nodes(K2,{0:"A",1:"B",2:"C",3:"D",4:"E"})
K_connector=nx.from_edgelist([(4,"X"),("X","A")])
```

```python
K=nx.compose_all([K1,K2,K_connector])
```

```python
nx.draw_spring(K,with_labels=True,node_color="pink")
plt.show()
```

```python
#all other nodes hav more degree centrality but betweenness centality will be more for x
print(nx.degree_centrality(K))
```

```python
print(nx.betweenness_centrality(K))
```
