```python
g6=nx.complete_graph(5)
g7=nx.complete_graph(5)
g7=nx.relabel_nodes(g7,{0:"A",1:"B",2:"C",3:"D",4:"E"})
G_connect=nx.from_edgelist([(4,"X"),("X","A")])
g8=nx.compose_all([g6,g7,G_connect])
```

```python
nx.draw_spring(g8,with_labels=True)
```

```python
print(nx.betweenness_centrality(g8))
```

```python
print(nx.degree_centrality(g8))
```

```python
#overall features
#density-edges/possible edges
print(nx.density(g8))
```

```python
#longest shortest path
print(nx.diameter(g8))
```
